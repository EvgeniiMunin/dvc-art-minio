```python
import mlflow
from pprint import pprint
#from pyngrok import ngrok

from sklearn.metrics import accuracy_score, f1_score, classification_report,  r2_score

import warnings
warnings.filterwarnings("ignore")
```

```python
def yield_artifacts(run_id, path=None):
    """Yield all artifacts in specified run"""
    client = mlflow.tracking.MlflowClient()
    for item in client.list_artifacts(run_id, path):
        if item.is_dir:
            yield from yield_artifacts(run_id, item.path)
        else:
            yield item.path
            
def fetch_logged_data(run_id):
    """Fetch params, metrics, tags and artifacts in specified run"""
    client = mlflow.tracking.MlflowClient()
    data = client.get_run(run_id).data
    # exclude system tags
    tags = {k: v for k, v in data.tags.items() if not k.startswith('mlflow.')}
    artifacts = list(yield_artifacts(run_id))
    return {
        "params": data.params,
        "metyrics": data.metrics,
        "tags": tags,
        "artifacts": artifacts
    }

def interpret_features(model, xtest):
    tempdf = pd.DataFrame(xtest, columns=feats).sample(frac=1)[:50]
    
    # create explainer to compute shap values
    explainer = shap.KernelExplainer(lambda x: model.predict(x), tempdf)
    
    # compute shap values for all xtest
    shap_values = explainer.shap_values(tempdf)
    
    fig, ax = plt.subplots(nrows=1, ncols=1)
    fig = shap.summary_plot(shap_values, tempdf, plot_size=(12, 8), show=False)
    plt.savefig(f'artifacts/shap.png')
    
def plot_preds(model, xtest):
    preds = model.predict(xtest)
    plt.figure(figsize=(15,5))
    plt.plot(preds_df['y_pred'], label='pred')
    plt.plot(preds_df['y_test'], label='actual')
    plt.legend()
    plt.savefig(f'artifacts/preds.png')
```

#### Abstract mlflow runs into method to test multiple models

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
def log_experiment_mlflow(model, run_name, xtrain, ytrain, xtest):
    with mlflow.start_run(run_name="RUN_{}".format(run_name)) as run:
        # start run and get status
        print("run_id: {}; status: {}".format(run.info.run_id, run.info.status))

        print(ytrain)
        
        model.fit(xtrain, ytrain)
        preds = model.predict(xtest)

        r2 = r2_score(preds, ytest)

        interpret_features(model, xtest)
        plot_preds(model, xtest)

        mlflow.log_metric(key='R2_val', value=r2)
        mlflow.log_artifact(local_path='artifacts/shap.png')
        mlflow.log_artifact(local_path='artifacts/preds.png')
        mlflow.sklearn.log_model(model, 'model_saved')
```

Lets make a function that trains different regression models and then logs the experiment results. We will log specifically
- R2 validation
- feature importance computed with Shap value
- predictions of the model
- model itself saved in .pkl

```python
model_list = [LinearRegression(), KNeighborsRegressor(), RandomForestRegressor()]
run_name_list = ['Linear Regression', 'KNeighbors Regressor', 'Random Forest Regressor']

for (model, run_name) in zip(model_list, run_name_list):
    log_experiment_mlflow(model, run_name, xtrain, ytrain, xtest)
```
